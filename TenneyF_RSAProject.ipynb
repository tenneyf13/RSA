```python
#Theorem 2: If n is a composite integer, then n has a prime divisor less than or equal to square 
#root of n. 
n = int(input("What is the n number?"))
def factorize(n):
    start = 3
    stop = n**0.5
    step = 2
    #The third parameter is known as a step which in this case will start with three then go up
    #by 2 till n-1 is reached. 
    for i in range(start,stop,step):
        if n%i == 0:
            return i 
        #Want to return same types. 
    return 0
p = factorize(n)
q = int(n / p)
print("Here are your two prime number {} and {}".format(p,q))
```
